# Home Value:

## Why?
[Studies](https://www.reinvestment.com/wp-content/uploads/2017/08/ReinvestmentFund_PA_District_Perf_Report.pdf) show a 1 percentage-point improvement in the share of students scoring “proficient or advanced” in reading or math was associated with roughly $513 to $620 increase in average home sales price across a district

[Studies](https://pmc.ncbi.nlm.nih.gov/articles/PMC5035645/?utm_source=chatgpt.com) show that there is a strong correlation with home values and neighborhood quality measures (such as poverty rate and observed property characteristics)

## Goal
Use house resale values to compare Pittsburgh neighborhoods. Metric (for now): Average sale price (last 3 years), valid sales only.
Dataset: https://data.wprdc.org/dataset/real-estate-sales

```python
# Imports
import pandas as pd
import matplotlib.pyplot as plt
from pathlib import Path
import matplotlib.ticker as mticker
```

```python
url = "https://data.wprdc.org/datastore/dump/5bbe6c55-bce6-4edb-9d04-68edeb6bf7b1"

# Read directly into a pandas DataFrame
df = pd.read_csv(url, low_memory=False)

print("Rows:", len(df))
df.head()
```

```python
# Clean data (filters out non-numeric prices, invalid dates, missing price/sale date/location, and outlier numbers that're suepr low/high)
work = df[['PRICE', 'SALEDATE', 'MUNIDESC']].copy()
work['PRICE'] = pd.to_numeric(work['PRICE'], errors='coerce')
work['SALEDATE'] = pd.to_datetime(work['SALEDATE'], errors='coerce')
work = work.dropna(subset=['PRICE', 'SALEDATE', 'MUNIDESC'])
work = work[(work['PRICE'] >= 1000) & (work['PRICE'] <= 5_000_000)]
print("Rows after clean:", len(work))
```

```python
# define a 3-year cutoff based on sale date
cutoff = pd.Timestamp.today() - pd.DateOffset(years=3)

# Keep only sales in the last 3 years
recent = work[work['SALEDATE'] >= cutoff]

# Calculate averages and format
avg_price = recent.groupby('MUNIDESC')['PRICE'].mean().sort_values(ascending=False)
avg_price_fmt = avg_price.round(0).map('{:,.0f}'.format)

print("Top 10 neighborhoods by average resale price (last 3 years):\n")
print(avg_price_fmt.head(10).to_string())
```

```python
# graph
ax = avg_price.head(10).plot(kind='bar', figsize=(10,5))
plt.title('Top 10 Areas by Average Resale Price')
plt.ylabel('Average Sale Price ($)')
plt.xticks(rotation=45, ha='right')
ax.yaxis.set_major_formatter(mticker.StrMethodFormatter('${x:,.0f}'))
plt.tight_layout()
plt.show()
```
